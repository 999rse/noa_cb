### Analyse data

```python
import matplotlib.pyplot as plt
import numpy as np

# Создайте случайные данные для примера (замените эту часть вашим тензором)
data = pars_smp_train

# Определите количество столбцов в гистограммах
num_bins = 20

# Создайте 5x3 подграфика
fig, axes = plt.subplots(3, 5, figsize=(15, 8))

for i in range(15):
    row = i // 5
    col = i % 5
    axes[row, col].hist(data[:, i, 0], bins=num_bins, edgecolor='k')
    axes[row, col].set_title(f'Параметр {i + 1}')
    axes[row, col].set_xlabel('Значения')
    axes[row, col].set_ylabel('Частота')
    axes[row, col].grid(False)

for i in range(15, 5*3):
    row = i // 5
    col = i % 5
    fig.delaxes(axes[row, col])

plt.tight_layout()
plt.show()
```

```python
# Test Shapiro-Wilk

data = pars_smp_train

for i in range(15):
    current_data = data[:, i, 0]
    
    stat, pvalue = stats.shapiro(current_data)
    
    if pvalue > 0.05:
        print(f'Параметр {i + 1}: Распределение близко к нормальному (pvalue = {pvalue:.4f})')
    else:
        print(f'Параметр {i + 1}: Распределение не является нормальным (pvalue = {pvalue:.4f})')

```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

data = pars_smp_train

fig, axes = plt.subplots(3, 5, figsize=(15, 8))

for i in range(15):
    current_data = data[:, i, 0]

    row = i // 5
    col = i % 5

    stats.probplot(current_data, dist="norm", plot=axes[row, col])

    axes[row, col].set_title(f'Q-Q график параметра {i + 1}')

plt.tight_layout()
plt.show()
```
